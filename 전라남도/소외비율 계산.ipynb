```python
import pandas as pd
import numpy as np
```

```python
buffer = pd.read_excel('후보군_고령인구/버퍼안_안전_휴일.xlsx')
```

```python
buffer = buffer[['gid','lbl','val','layer']]
```

```python
candidate = pd.read_csv('후보군_고령인구/후보군_고령인구.csv',encoding='cp949')
```

```python
city = ['완도군','장성군','함평군','영암군','강진군','진도군','구례군','담양군','장흥군','신안군']

for i in city :
    df = candidate[candidate['layer'].str.contains(i)]
    df_ = buffer[buffer['layer'].str.contains(i)]
    df.drop_duplicates(subset=None,
                    keep='first',
                    inplace=True,
                    ignore_index=False)
    df_.drop_duplicates(subset=None,
                    keep='first',
                    inplace=True,
                    ignore_index=False)
    print(i, (df['val'].sum() - df_['val'].sum()) / df['val'].sum())
```
