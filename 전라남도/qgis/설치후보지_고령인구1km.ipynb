```python
import pandas as pd
import numpy as np
```

```python
pop1 = pd.read_csv('qgis/최종 1 고령인구.csv',encoding='cp949')
pop3 = pd.read_csv('qgis/최종 3 고령인구.csv',encoding='cp949')
```

```python
layer = list(pop1['layer'].unique())
layer
```

```python
pop1['layer_2'].unique()
```

```python
layer[0]
```

```python
pop1_drugstore = pop1[pop1['layer']==layer[0]]
pop1_welfare_DamYang = pop1[pop1['layer']==layer[1]]
pop1_healthB = pop1[pop1['layer']==layer[2]]
pop1_healthC = pop1[pop1['layer']==layer[3]]
pop1_welfare_HamPyeong = pop1[pop1['layer']==layer[4]]
pop1_welfare_ShinAhn = pop1[pop1['layer']==layer[5]]
pop1_healthA = pop1[pop1['layer']==layer[6]]
```

```python
pop1_drugstore = pop1_drugstore[['약국명_y','layer_2','좌표정보(X)','좌표정보(Y)','gid','val']]
pop1_drugstore.rename(columns={'약국명_y':'name',
                               '좌표정보(X)':'Longitude',
                               '좌표정보(Y)':'Latitude'}
                      ,inplace=True)
pop1_drugstore
```

```python
pop1_welfare_DamYang = pop1_welfare_DamYang[['회관명','layer_2','Longitude','Latitude','gid','val']]
pop1_welfare_DamYang.rename(columns={'회관명':'name'},inplace=True)
pop1_welfare_DamYang
```

```python
pop1_welfare_DamYang.info()
```

```python
pop1_welfare_HamPyeong = pop1_welfare_HamPyeong[['경로당명','layer_2','Longitude','Latitude','gid','val']]
pop1_welfare_HamPyeong.rename(columns={'경로당명':'name'},inplace=True)
pop1_welfare_HamPyeong
```

```python
pop1_welfare_ShinAhn = pop1_welfare_ShinAhn[['마을회관명','layer_2','Longitude','Latitude','gid','val']]
pop1_welfare_ShinAhn.rename(columns={'마을회관명':'name'},inplace=True)
pop1_welfare_ShinAhn
```

```python
pop1_healthA = pop1_healthA[['보건소명','layer_2','Longitude','Latitude','gid','val']]
pop1_healthB = pop1_healthB[['기관명','layer_2','Longitude','Latitude','gid','val']]
pop1_healthC = pop1_healthC[['기관명','layer_2','Longitude','Latitude','gid','val']]
pop1_healthA.rename(columns={'보건소명':'name'},inplace=True)
pop1_healthB.rename(columns={'기관명':'name'},inplace=True)
pop1_healthC.rename(columns={'기관명':'name'},inplace=True)
```

```python
pop1 = pd.concat([pop1_drugstore, pop1_welfare_DamYang,
                  pop1_welfare_ShinAhn,pop1_welfare_HamPyeong,
                  pop1_healthA,pop1_healthB,pop1_healthC],
                 axis=0,
                 ignore_index = True)
```

```python
pop1.drop('index',axis=1,inplace=True)
pop1.keys()
```

```python
pop1['index'] = pop1['name'] + str(pop1['Longitude'])
```

```python
pop1
```

```python
pop1['index'] = str(pop1['Latitude']) + str(pop1['Longitude'])
```

```python
pop1_ = pop1.groupby('name').sum('val')
```
