```python
#칼럼명 한글로 되어있는 부분 영어로 변경(코딩 용이성을 위해)
b3 = b.rename(columns = {'월요일 운영':'monday'})
b3
```

```python
#~를 기준으로 운영시간 나눔
b3['monday1'] = b3.monday.str.split('~').str[0]
b3['monday2'] = b3.monday.str.split('~').str[1]
```

```python
b3.head()
```

```python
#:를 기준으로 시간만 표시될 수 있도록 나눔
b3['monday_open'] = b3.monday1.str.split(':').str[0]
b3['monday_close'] = b3.monday2.str.split(':').str[0]
b3
```
